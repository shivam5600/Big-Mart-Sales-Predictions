## For missing values

```python
df1 = df[df['Item_Weight']<8] 
df2 = df[(df['Item_Weight']>8) & (df['Item_Weight']<15) ] 
df3 = df[df['Item_Weight']>16] 

```

```python
df1['Item_Weight'] = df1['Item_Weight'].fillna(6)
df2['Item_Weight'] = df2['Item_Weight'].fillna(12)
df3['Item_Weight'] = df3['Item_Weight'].fillna(18)
```

```python
df_ = pd.concat([df1,df2,df3],axis='rows')
```

```python
df.shape
```

```python
df_.isnull().sum()
```

```python
df1 = df_[df_['Item_Outlet_Sales']<1800] 
df2 = df_[(df_['Item_Outlet_Sales']>1800) & (df_['Item_Outlet_Sales']<5000) ] 
df3 = df_[df_['Item_Outlet_Sales']>5000] 

```

```python
df1['Outlet_Size'] = df1['Outlet_Size'].fillna('Small')
df2['Outlet_Size'] = df2['Outlet_Size'].fillna('Medium')
df3['Outlet_Size'] = df3['Outlet_Size'].fillna('High')
```

```python
df_final = pd.concat([df1,df2,df3],axis='rows')
```

```python
df_final.reset_index()
```

```python
df_final['Outlet_Size']= df_final['Outlet_Size'].str.lower()
```

```python
df_final.reset_index(inplace=True)
```
